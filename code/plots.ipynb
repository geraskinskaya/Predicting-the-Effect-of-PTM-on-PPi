```python
import numpy as np
import matplotlib.pyplot as plt

# Data values
TabNet = [ 0.79, 0.79, 0.79, 0.79, 0.54, 0.90]
XGBoost = [ 0.84, 0.84, 0.84, 0.84, 0.65, 0.96]
LightGBM = [0.84, 0.85, 0.84, 0.85, 0.65, 0.97]
RandomForests = [  0.84, 0.84, 0.84, 0.84, 0.64, 0.96]

# Bar width
barWidth = 0.2

# Positions for bars
br1 = np.arange(len(TabNet))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

# Adding a gap between groups of four bars
gap = 0.3
br1 = [x + i*gap for i, x in enumerate(br1)]
br2 = [x + i*gap for i, x in enumerate(br2)]
br3 = [x + i*gap for i, x in enumerate(br3)]
br4 = [x + i*gap for i, x in enumerate(br4)]

# Colors extracted from the image
colors_chart_normalized = [
    (133/255, 164/255, 210/255),
    (234/255, 177/255, 90/255),
    (232/255, 143/255, 232/255),
    (221/255, 219/255, 216/255)
]

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(12, 8))
plt.style.use('seaborn-v0_8-pastel')
ax.patch.set_facecolor('white')

plt.bar(br1, TabNet, width=barWidth,label='TabNet')
plt.bar(br2, XGBoost,  width=barWidth,  label='XGBoost')
plt.bar(br3, LightGBM,  width=barWidth,  label='LightGBM')
plt.bar(br4, RandomForests, width=barWidth,  label='Random Forests')
# X-tick positions
centered_xtick_positions = [r + 1.3*barWidth + 0.1*i*gap for i, r in enumerate(br1)]
plt.xticks(centered_xtick_positions, ['Accuracy', 'Precision', 'Recall', 'F1', 'MCC', 'ROC-AUC'])

plt.xlabel('Metrics', fontweight='bold', fontsize=15)
plt.ylabel('Performance', fontweight='bold', fontsize=15)
plt.legend()
plt.show()
```
